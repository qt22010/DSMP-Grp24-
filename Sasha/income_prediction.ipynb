```python
#importing packages 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/sasha/Desktop/DATASCIENCE/2023/Mini-Project/data.nosync/cleaned_EXPENDITURE_data.csv')
df = df.drop(['Unnamed: 0'], axis =1)
df.head()
df = df.rename(columns={'type':'type_'})
```

```python
customers = df.groupby(['from_account','type_']).agg({'money_amount':'sum'})
customers = customers.reset_index().pivot(index = 'from_account', columns = 'type_', values='money_amount')
customers['x'] = customers['Inferior ']+(2*customers['Normal '])
customers = customers.reset_index()
customers.head()
```

```python
percentiles = pd.read_csv('/Users/sasha/Desktop/DATASCIENCE/2023/Mini-Project/data.nosync/Percentiles_Income.csv')
percentiles['After_tax'] = percentiles['After_tax'].map(lambda x: int(x.replace(',','')))
percentiles['Before_tax'] = percentiles['Before_tax'].map(lambda x: int(x.replace(',','')))
percentiles.head()
```

```python
x_s = np.array(customers['x'])
from scipy import stats
customers['Percentile'] = customers['x'].map(lambda x: round(stats.percentileofscore(x_s, x)))
customers['Percentile'] = customers['Percentile'].map(lambda x: 1 if x==0 else 99 if x==100 else x)
customers = pd.merge(customers, percentiles)
customers.head()
```
