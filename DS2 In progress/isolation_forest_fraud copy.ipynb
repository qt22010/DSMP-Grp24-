```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/Users/sasha/Desktop/DATASCIENCE/2023/Mini-Project/data.nosync/clean_data_set_2.csv", dtype={'Account Number':str, 'Third Party Account Number':str}, index_col=0)
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()  
transformed = ohe.fit_transform(df[['to_name']])
df[ohe.categories_[0]] = transformed.toarray()
```

```python
df.columns
```

```python
from sklearn.ensemble import IsolationForest

customer_list  = df.Account.unique()
df_fraud = df.drop(['Account', 'to_account', 'to_name', 'Date', 'Time', 'transaction_ID'], axis=1)
for x in customer_list:
    temp = df_fraud[df.Account == x]
    model=IsolationForest(n_estimators=50, max_samples='auto', contamination=float(0.1),max_features=1.0)
    model.fit(temp) 
    df.loc[df.Account == x,'scores']=model.decision_function(temp)
    df.loc[df.Account == x,'anomaly']=model.predict(temp)
```

```python
anomaly=df.loc[df['anomaly']==-1]
anomaly_index=list(anomaly.index)
print(anomaly)
```
